```python
class Character():
  def __init__(self, d = {'name': None, 'stats':[0,0,0,0,0,0,]}):
    self.name = d['name']
    self.stats = d['stats']
  def get_stat(self, stat):
    stats = ['str','dex','con','int','wis','cha']
    if type(stat) == int:
      return self.stats[stat]
    elif type(stat) == str:
      return self.stats[stats.index(stat)]
    else:
        None

moted = Character({'name': "Moted", "stats": [100, 80, 80, 60, 80, 80]})
null = Character()
print(null.name)
print(f'Moted:  {moted.name}')
null.name = "Null"
print(null.name)
print(moted.name)
```
