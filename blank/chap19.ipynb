```python
class Character():
  """ Base character class """


  def __init__(self, d = {'name': None, 'race': 'Human', 'class': None,  'stats': [0,0,0,0,0,0]}):
    self.name = d['name']
    self.stats = d['stats']
    self.race = d['race']
    self.cls = d['class']# add class


  def get_stat(self, stat):
    stats = ['str', 'dex', 'con', 'int', 'wis', 'cha']
    if type(stat) == int:
      return self.stats[stat]
    elif type(stat) == str:
      return self.stats[stats.index(stat)]
    else:
      None


  def pcs(self):
    print(f'Character sheet for {self.name}\n')
    print(f'Race: {self.race}')
    print(f"Class: {self.cls}")  ## print class
    print(f'Stats:')
    print(f'Strength:     {self.get_stat('str')}   Intelligence: {self.get_stat('int')}')
    print(f'Dexterity:    {self.get_stat('dex')}    Wisdom:       {self.get_stat('wis')}')
    print(f'Constitution: {self.get_stat('con')}    Charisma:     {self.get_stat('cha')}')

moted = Character({'name': "Bilbo Baggens", 'race': "Hobbit",'class': "Burgular",'stats': [100, 80, 80, 60, 80, 80]})
#moted = Character({'name': "Moted", 'race': 'Barbarian', "stats": [100, 80, 80, 60, 80, 80]})
moted.pcs()
```
