```python

class Race():
    def __init__(self, d = {
        'stats_modifier': [1, 1, 1, 1, 1, 1],
        'innate_abilities': [],
        'size': 'medium'}):

        self.stats_modifiers = d['stats_modifier']
        self.innate_abilities = d['innate_abilities']
        self.size = d['size']

    def get_mod(self, stat):
        stats = ['str', 'dex', 'con', 'int', 'wis', 'cha']
        if isinstance(stat, int):
            return self.stats_modifiers[stat]
        elif isinstance(stat, str):
            return self.stats_modifiers[stats.index(stat)]
        else:
            return None

class Character():
    """ Base character class """

    def __init__(self, d = {
        'name': None,
        'race': 'Human',
        'class': 'Warrior',
        'level': 1,
        'stats': [0,0,0,0,0,0]}):

        self.name = d['name']
        self.stats = d['stats']
        self.race = d['race']
        self.level = d['level']
        self.cclass = d['class']

    def __str__(self):
        return f"{self.name} ({self.race} {self.cclass}, Level {self.level})"
    def get_stat(self, stat):
        stats = ['str', 'dex', 'con', 'int', 'wis', 'cha']
        if isinstance(stat, int):
            return self.stats[stat]
        elif isinstance(stat, str):
            return self.stats[stats.index(stat)]
        else:
            return None

    def computed_stat(self, stat):
        return int(self.get_stat(stat) * races[self.race].get_mod(stat))
    def pcs(self):
        print(f"Character sheet for {self.name}\n")
        print(f"Race: {self.race}")
        print(f"Class: {self.cclass}")
        print(f"Level: {self.level}")
        print(f"Stats:")
        print(f"Strength:     {self.get_stat('str'):>3}    Intelligence: {self.get_stat('int'):>3}")
        print(f"Dexterity:    {self.get_stat('dex'):>3}    Wisdom:       {self.get_stat('wis'):>3}")
        print(f"Constitution: {self.get_stat('con'):>3}    Charisma:     {self.get_stat('cha'):>3}")
        print("\n")

class Item():
    def __init__(self, d = {
        'name': None,
        'description': None}):

        self.name = d['name']
        self.description = d['description']



races = {}

races['Human'] = Race()
races['Dwarf'] = Race({
    'stats_modifier': [1.1, 1, 1.3, 0.9, 1, 1],
    'size': 'medium_small',
    'innate_abilities': ['infravision']
})
races['Grey Elf'] = Race({
    'stats_modifier': [0.7, 1.2, 0.8, 1.1, 1, 1.1],
    'size': 'medium_small',
    'innate_abilities': ['infravision', 'outdoor_sneak']
})
races['Ogre'] = Race({
    'stats_modifier': [1.5, 0.8, 1.5, 0.5, 0.75, 0.75],
    'size': 'large',
    'innate_abilities': ['doorbash']
})
characters = {}
characters['Moted'] = Character({
    'name': "Moted",
    'race': 'Dwarf',
    'class': 'Shaman',
    'level': 24,
    'stats': [88, 80, 80, 80, 80, 80]
})
characters['Aleolas'] = Character({
    'name': "Aleolas",
    'race': 'Grey Elf',
    'class': 'Ranger',
    'level': 50,
    'stats': [100, 80, 100, 80, 80, 80]
})
characters['Illilel'] = Character({
    'name': "Illilel",
    'race': "Grey Elf",
    'class': "Bard",
    'level': 50,
    'stats': [100, 100, 100, 72, 54, 100]
})
dagger = Item({
    'name': "Dagger of Oblivion",
    'description': "A stabby thing"
})
def character_list(char_list):
    print(f'{"Name":<15} {"Race":<12} {"Class":<10} {"Level":>5}')
    print("_" * 45, "\n")

    for char in char_list.values():
        print(f'{char.name:<15} {char.race:<12} {char.cclass:<10} {char.level:>5}')
character_list(characters)
print("\n--- Character Sheets ---\n")
characters['Moted'].pcs()
print("Computed Strength (Moted):", characters['Moted'].computed_stat('str'))
```
