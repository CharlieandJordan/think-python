```python
## Base Race Class
class Race():
  def __init__(self, d = {
      'stats_modifier': [1, 1, 1, 1, 1, 1],
      'innate_abilities': [],
      'size': 'medium'}):
    self.stats_modifiers = d['stats_modifier']

  def get_mod(self, stat):
    stats = ['str', 'dex', 'con', 'int', 'wis', 'cha']
    if type(stat) == int:
      return self.stats_modifiers[stat]
    elif type(stat) == str:
      return self.stats_modifiers[stats.index(stat)]
    else:
      None
```

```python
terrain = ('no ground', 'water', 'mountain', 'plain', 'stone', 'wood floor')

class Room():
  def __init__(self, d):
    self.number = d['number']
    self.name = d['name']
    self.description = d['description']
    self.indoors = d['indoors']
    self.terrain = d['terrain']
    self.exits = d['exits']
    self.objects = d['objects']
    self.characters = d['characters']

  def _exits(self):
    output = "\nExits:"
    if len(self.exits) == 0:
      output += " None!"
    else:
      count = 0
      for exit in self.exits:
        if count == 1:
          output += ","
        output += f' {exit['direction'].title()}'
        count += 1
    return output + "\n\n"

  def _objects(self):
    output = ""
    if len(self.objects) > 0:
      for obj in self.objects:
        output += f'{obj.room_description}\n'
    return output

  def _characters(self):
    output = ""
    if len(self.characters) > 0:
      for character in self.characters:
        output += f'{characters[character].name.title()} stands here\n'
    return output

  def __repr__(self):
    output = f'{self.name}\n     {self.description}\n'
    output += self._exits()
    output += self._objects()
    output += self._characters()
    return output
```
